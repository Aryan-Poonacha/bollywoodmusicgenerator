```python
#ATTEMPT 1 TO READ MIDI FILES


from music21 import converter, instrument, note, chord
import os
import pickle

#roughly same type of midi extraction process in:
#https://github.com/DebarghaG/Tansen/blob/master/train_network.py#L22
#https://hedonistrh.github.io/2018-04-27-Music-Generation-with-LSTM/
#https://medium.com/@leesurkis/how-to-generate-techno-music-using-deep-learning-17c06910e1b3
def parse_music():
    
    notes = []
    
    for file in os.listdir('data'):
        
        
        
        print('parsing ' + file + '...')
        midi_file = converter.parse('data/' + file)
        notes_to_parse = None
        
        instrument_data = None
        try:
            print('contains instruments!')
            instrument_data = instrument.partitionByInstrument(midi_file)
            print(instrument_data)
            notes_to_parse = instrument_data.parts[0].recurse()
        except:
            notes_to_parse = midi_file.flat.notes
        #print(notes_to_parse)
        
        instrument_names = []
        try:
            for instrument in instrument_data: # Learn names of instruments.
                name = (str(instrument).split(' ')[-1])[:-1]
                instrument_names.append(name)
            print(instrument_names)
        except TypeError:
            print ('Type is not iterable.')
        
        
        #extracts all the features sequentially from the MIDI file
        components = []
        for element in midi_file.recurse():
                components.append(element)
        
        
        for instrument_element in notes_to_parse:
            #print(instrument_element)
            if isinstance(instrument_element, note.Note):
                notes.append(str(element.pitch))
            elif isinstance(instrument_element, chord.Chord):
                notes.append('.'.join(str(n) for n in element.normalOrder))
    
    with open('notes', 'wb') as filepath:
        pickle.dump(notes, filepath)
        
    return notes

print(parse_music())
```
